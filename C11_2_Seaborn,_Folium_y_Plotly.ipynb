#Seaborn

```python
import seaborn as sns
```

```python
#Grafico de lineas
#Datos de ejemplo
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 5, 3]
y2 = [1, 3, 2, 4, 5]
#Creamos grfico
sns.lineplot(x=x, y=y1, label='Lnea 1', marker='o', linestyle='-', markersize=10)
sns.scatterplot(x=x, y=y1, color='blue', s=100, label='_nolegend_')
sns.lineplot(x=x, y=y2, label='Lnea 2', marker='s', linestyle='--', markersize=10)
sns.scatterplot(x=x, y=y2, color='orange', s=100, label='_nolegend_')
plt.title('Grfico de Lneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.show()
```

```python
#Cargamos un conjunto de datos de ejemplo de Seaborn
datos_iris = sns.load_dataset("iris")

#Creamos un pair plot con Seaborn
sns.pairplot(datos_iris, hue="species", markers=["o", "s", "D"])
plt.show()
```

#Folium

```python
import folium
```

```python
#Coordenadas
latitud = -33.4489
longitud = -70.6693
#Mapa
mapa = folium.Map(location=[latitud, longitud])
mapa
```

```python
#Coordenadas
latitud = -33.4489
longitud = -70.6693
#Mapa
mapa = folium.Map(location=[latitud, longitud], zoom_start=12)
#Marcador para Santiago de Chile
folium.Marker([latitud, longitud], popup="Santiago de Chile").add_to(mapa)
#Capa de informacin de trfico
folium.TileLayer('Stamen Terrain').add_to(mapa)
#Marcador adicional
folium.Marker([-33.4372, -70.6506], popup="Otro Marcador").add_to(mapa)
#Crculo alrededor del segundo marcador
folium.CircleMarker(location=[-33.4372, -70.6506], radius=50, color='red', fill=True, fill_color='red').add_to(mapa)
# Mostrar el mapa
mapa.save("mapa_santiago_completo.html")
```

#Plotly

```python
import plotly.express as px
```

```python
#Datos de ejemplo
datos = px.data.tips()
#Grfico
fig = px.box(datos, x="day", y="total_bill", points="all", title="Boxplot de Total Bill por Da")
#Mostrar el grfico
fig.show()
```

```python
#Datos de ejemplo
datos = px.data.tips()
#Creamos grfico con diversos parametros
fig = px.box(datos,
             x="day",
             y="total_bill",
             points="all",
             title="Boxplot de Total Bill por Da",
             color="sex",  # Agregar color por gnero
             hover_data=["smoker"],  # Informacin adicional al pasar el mouse
             notched=True,  # Agregar muescas en los boxplots
             labels={"total_bill": "Monto Total de la Cuenta"},  # Cambiar etiquetas de ejes
             boxmode="group"  # Agrupar los boxplots
             )
#Otras personalizaciones
fig.update_layout(
    yaxis=dict(title='Monto Total de la Cuenta'),
    boxgap=0.1,  # Espaciado entre los boxplots
    boxgroupgap=0.05  # Espaciado entre los grupos de boxplots
)
# Mostrar el grfico
fig.show()
```

```python
#Datos de ejemplo
datos = px.data.gapminder()
#Filtro
datos_2007 = datos[datos['year'] == 2007]
#Grfico de burbujas en 2D
fig = px.scatter(datos_2007,
                 x='gdpPercap',
                 y='lifeExp',
                 size='pop',  # Tamao de las burbujas basado en la poblacin
                 color='continent',  # Color por continente
                 hover_name='country',  # Informacin adicional al pasar el mouse
                 title='Grfico de Burbujas en 2D (Ao 2007)')

#Personalizar diseo
fig.update_layout(
    xaxis=dict(title='PIB per cpita'),
    yaxis=dict(title='Expectativa de Vida'),
    showlegend=True  # Mostrar leyenda
)
#Mostrar el grfico
fig.show()
```

```python
#Datos de ejemplo
datos = px.data.gapminder()
#Filtro
datos_2007 = datos[datos['year'] == 2007]
#Grfico de burbujas en 3D
fig = px.scatter_3d(datos_2007,
                    x='gdpPercap',
                    y='lifeExp',
                    z='pop',  # Eje z: Poblacin
                    size='pop',  # Tamao de las burbujas basado en la poblacin
                    color='continent',  # Color por continente
                    hover_name='country',  # Informacin adicional al pasar el mouse
                    title='Grfico de Burbujas en 3D (Ao 2007)')
#Personalizar
fig.update_layout(
    scene=dict(
        xaxis=dict(title='PIB per cpita'),
        yaxis=dict(title='Expectativa de Vida'),
        zaxis=dict(title='Poblacin')
    )
)
#Mostrar el grfico
fig.show()
```
